```python
x_pic,y_pic = flag_pic.size
pic = asarray(flag_pic)
new_pic = zeros((x_pic,y_pic))
for x in range(x_pic):
    for y in range(y_pic):
        if pic[y][x][3] != 0:
            print("\tget_area({x},{y})".format(y = y,x = x))
```

```python
def the_angle(data,deg):
    if data.imag >= 0:
        return angle(data,deg)
    else:
        return 360 - angle(complex(data.real,-data.imag),deg)
        
```

```python
complex_data = []
real_x = []
imag_y = []
theta_data = []
for x in range(x_pic):
    for y in range(y_pic):
        if pic[y][x][3] != 0:
            complex_data.append(complex(x,y))
            real_x.append(x - x_pic // 2)
            imag_y.append(-y + y_pic // 2) 
            theta_data.append(the_angle(complex(x - x_pic // 2, -y + y_pic // 2),True))
```

```python
def read_pic(path):
    # 获取图像矩阵
    pic = img.open(path)
    pixel = pic.load()
    # 以rgba格式分别区分
    A = zeros((pic.size))
    for x in range(pic.size[0]):
        for y in range(pic.size[1]):
            A[x][y] = pixel[x,y][3] // 255
    return A
data = read_pic('flag.png')
print(data)
```
